BOW (Bag of words)

```python
from sklearn.feature_extraction.text import CountVectorizer

print("\n===== BAG OF WORDS ======")
bow_vectorizer=CountVectorizer()
corpus = ['This is the first sentence', 'This is the second sentence']
x_bow=bow_vectorizer.fit_transform(corpus)
print("vocabulary:",bow_vectorizer.vocabulary_)
print("BOW Matrix:",x_bow.toarray())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

print("\n===== TF-IDF ======")
tfidf_vectorizer = TfidfVectorizer()
# Assuming 'corpus' is defined from the previous cell
x_tfidf = tfidf_vectorizer.fit_transform(corpus)
print("vocabulary:", tfidf_vectorizer.vocabulary_)
print("TF-IDF Matrix:", x_tfidf.toarray())
```

### Explanation of TF-IDF Matrix and Vocabulary

The output from the `TfidfVectorizer` consists of two main parts:

*   **Vocabulary:** The `vocabulary_` attribute is a dictionary that maps each unique word found in the corpus to an index. This index represents the column that corresponds to that word in the TF-IDF matrix.

    For example, in the output:
    `{'this': 5, 'is': 1, 'the': 4, 'first': 0, 'sentence': 3, 'second': 2}`
    The word 'this' is at index 5, 'is' is at index 1, and so on.

*   **TF-IDF Matrix:** The `TF-IDF Matrix` is a numerical representation of your text data where each row represents a document (or sentence in this case) and each column represents a word from the vocabulary. The value in each cell is the TF-IDF score for that word in that document.

    The TF-IDF score is calculated as:

    **TF-IDF = TF (Term Frequency) * IDF (Inverse Document Frequency)**

    *   **Term Frequency (TF):** How often a word appears in a specific document.
    *   **Inverse Document Frequency (IDF):** A measure of how important or unique a word is across the *entire* set of documents. Words that appear in many documents have a lower IDF score, while words that are rare have a higher IDF score.

    The TF-IDF score highlights words that are frequent in a particular document but not very common in other documents in the corpus.

```python
from sklearn.feature_extraction.text import CountVectorizer

print("\n===== N-GRAMS ======")
# Define the corpus (list of sentences)
corpus = ['This is the first sentence', 'This is the second sentence','This is not second sentence']

# Initialize CountVectorizer for N-grams.
# ngram_range=(1, 1) for unigrams (single words)
# ngram_range=(2, 2) for bigrams (two-word phrases)
# ngram_range=(3, 3) for trigrams (three-word phrases)
# ngram_range=(1, 2) for unigrams and bigrams
ngram_vectorizer = CountVectorizer(ngram_range=(2, 2)) # Example: using bigrams

# Fit the vectorizer to the corpus and transform the corpus into a matrix of N-gram counts
x_ngrams = ngram_vectorizer.fit_transform(corpus)

# Print the vocabulary (mapping of N-grams to indices)
print("Vocabulary:", ngram_vectorizer.vocabulary_)

# Print the N-gram matrix (count of each N-gram in each sentence)
print("N-gram Matrix:\n", x_ngrams.toarray())
```
